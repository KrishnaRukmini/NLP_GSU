1. Using NLTK tokenize all documents, separated by polarity, remove stop words, and list the top 20 most frequent tokens (and their counts) for the positive reviews, and the top
20 most frequent tokens (and their counts). What kind of things do you notice are
different between the two sets? (30 points)

```
import os

data_dir = "/content/txt_sentoken"
files = os.listdir(data_dir+"/")
files
```

```
def List_Of_Files(dir_Name): 
    listOfFile = [x for x in os.walk(dir_Name)]
    return listOfFile   
```

```
count = 0
for fl in files:
  if count == 1  :
    polarity_positive = List_Of_Files("/content/txt_sentoken/" + fl +"/")
  else:
    #print(file_train)
    polarity_negative = List_Of_Files("/content/txt_sentoken/" + fl +"/")
    count =+ 1
```

```
data_positive = []
for p_p in polarity_positive[0][2]:
  #print (p_p[2])
  with open(polarity_positive[0][0]+p_p,'r') as fs:
    data_positive.append(fs.read())
```

```
data_negative = []
for p_n in polarity_negative[0][2]:
  #print (p_p[2])
  with open(polarity_negative[0][0]+p_n,'r') as fs:
    data_negative.append(fs.read())
```

```
import pandas as pd

positive_df = pd.DataFrame(data_positive,columns = ['Data'])
positive_df['polarity'] = 1

negative_df = pd.DataFrame(data_negative,columns = ['Data'])
negative_df['polarity'] = 0
```

```
positive_df
```

```
negative_df
```

```
data_df = pd.concat([positive_df,negative_df])
```

```
data_df
```

```
data_df.polarity.value_counts()
```

```
positive_reviews = "".join(map(str, positive_df['Data']))
```

```
negative_reviews = "".join(map(str, negative_df['Data']))
```

```
import nltk
nltk.download('punkt')
```

```
tokens_pr = nltk.word_tokenize(positive_reviews.lower())
tokens_nr = nltk.word_tokenize(negative_reviews.lower())
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')


stop_words = stopwords.words('english')
```

```
remove_sw_pr = [word for word in tokens_pr if not word in stop_words and word.isalpha()]
remove_sw_nr = [word for word in tokens_nr if not word in stop_words and word.isalpha()]
```

```
len(remove_sw_nr)
```

```
 tokens_nr_freqdist = nltk.FreqDist(remove_sw_nr)
 tokens_pr_freqdist = nltk.FreqDist(remove_sw_pr)
```

```
for word, frequency in tokens_nr_freqdist.most_common(20):
    print('%s  ==>  %d' % (word, frequency))
```

```
for word, frequency in tokens_pr_freqdist.most_common(20):
    print('%s  ==>  %d' % (word, frequency))
```

```
wordfreq = []
for w in remove_sw_pr:
    wordfreq.append(remove_sw_pr.count(w))
```

```
wordfreq_nr = []
for w in remove_sw_nr:
    wordfreq_nr.append(remove_sw_nr.count(w))
```

```
Pairs_pr = list(set(zip(remove_sw_pr, wordfreq)))
```

```
Pairs_nr = list(set(zip(remove_sw_nr, wordfreq)))
```

```
def Sort_Tuple(tup):   
    tup.sort(key = lambda x: x[1])  
    return tup  

Pairs_pr_sorted = (Sort_Tuple(Pairs_pr))
pairs_nr_sorted = (Sort_Tuple(Pairs_nr))
```
