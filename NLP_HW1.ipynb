4. Write a regular expression using groups in Python that extracts all phrases that have
“son of ” and extract/output the answer in a case-insensitive way. For example, for the
following “I am the son of earth”, should find “son of earth” and extract/output “earth”.


```
r = re.compile(r'(.*) son of (.*)')
```

```
for line in text.split("\n"):
    m = r.search(line, re.I)
    if m:
        print(m.group(2))
```

5. Plays are divided in acts, write a regular expression using groups in Python that
extracts the number of each act found in this play in a case-sensitive way

```
r = re.compile(r'^ACT (.*)')
for line in text.split("\n"):
    m = r.search(line)
    if m:
        print(m.group(1))
```

6. Write a Python function that takes the file as an input and outputs how many lines end
with: ? , that is a question mark at the end of the line. This function needs to use a
regular expression.

```
def sample_func(f_name):
    with open(f_name, 'r') as fp:
        txt = fp.read()
    return len(re.findall(r'\?$', txt, re.M))
```

7.Using the code from Class 04, train a Random Forest classification model for the same
two newsgroups, generate the confusion matrix, and all the same performance metrics

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```
categories = ['rec.autos','comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
```

```
print(train.data[6])
```

```
from sklearn.ensemble import RandomForestClassifier 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import make_pipeline
import pandas as pd 
```

```
model_RFC = make_pipeline(CountVectorizer(), RandomForestClassifier())
model_RFC.fit(train.data, train.target)
labels_RFC = model_RFC.predict(test.data)
```

```
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels_RFC)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```
import sklearn
print('Accuracy:', sklearn.metrics.accuracy_score(test.target,labels_RFC))
print('Precision:', sklearn.metrics.precision_score(test.target,labels_RFC))
print('Recall:', sklearn.metrics.recall_score(test.target,labels_RFC))
print('F1 Score:', sklearn.metrics.f1_score(labels_RFC, test.target, average='macro'))
print('Mean Absolute Error:', sklearn.metrics.mean_absolute_error(test.target,labels_RFC))
print('Mean Squared Error:', sklearn.metrics.mean_squared_error(test.target, labels_RFC))  
print('Root Mean Squared Error:', np.sqrt(sklearn.metrics.mean_squared_error(test.target, labels_RFC)))
```
